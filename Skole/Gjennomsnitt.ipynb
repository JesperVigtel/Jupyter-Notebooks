# Gjennomsnitt

```python
tall = [1,3,1,3]

def gjennomsnitt(verdier):
    n = len(verdier)
    summen = 0
    for i in range(n):# summen = sum(verdier) (litt juks)
        summen = summen + verdier[i]
    snitt = summen/n
    return snitt

#kan ogsp skrives som

def gjennomsnitt_1(verdier):
    n = len(verdier)
    summen = 0
    for verdi in verdier:
        summen = summen + verdi
    snitt = summen/n
    return snitt
    
snitt = gjennomsnitt(tall)
print(snitt)
```

```python
def standaravvik(verdier):
    n = len(verider)
    snitt = gjennomsnitt(verdier)
    avvik = 0
    for verdi in verdier:
        avvik = avvik + (verdi - snitt)**2
    avvik = (avvik/n)**0.5
    return avvik
    
```

# Regresjon

```python
from pylab import *

data = loadtxt("temperatur.txt",skiprows = 1)

tid = data[:,0]
temp = data [:,1]

plot(tid,temp,color = "hotpink")
xlabel("tid (min)")
ylabel("Temperatur (C)")
show()
```

```python

reg = polyfit(tid, temp,2)
a = reg[0]
b = reg[1]
c = reg[2]
```

```python
a , b,c,d = polyfit(tid, temp,3)

def modell(x):
    return a*x**3+b*x**2+c*x+d

x  =linspace(0,350,10000)
y = modell(x)
plot(x,y, label = "Tilpasset kurve")
plot(tid, temp, linestyle = "--", label = "Datapunkter")
legend()
show()
```
