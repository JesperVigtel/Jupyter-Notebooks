$$ f'(x) = \frac{f(x+\Delta x)-f(x)}{(\Delta x)}$$

```python
from pylab import *

a = 0        # Startverdi
b = 5        # Sluttverdi
dx = 1E-5   # Steglengde
N = int((b-a)/dx) # Antall intervaller
y0 = 1       # Initialbetingelse

# Matriser
x = zeros(N+1)
y = zeros(N+1)
y[0] = y0

# Den deriverte av funksjonen y eller f(x)
def yder(y,x):
  return 1
  
# Eulers metode
for i in range (N):
  y[i+1] = y[i] + yder(y[i],x[i])*dx
  x[i+1] = x[i] + dx
  
plot(x,y)
grid()
  
```

```python
from pylab import *

a = 0             # Startverdi
b = 20            # Sluttverdi
dx = 1E-5         # Steglengde
N = int((b-a)/dx) # Antall intervaller
y0 = 100          # Initialbetingelse
k = -0.5
T_omg = 20

# Matriser
x = zeros(N+1)
y = zeros(N+1)
y[0] = y0


# Den deriverte av funksjonen y eller f(x)
def yder(y,x):
  return k*(y-T_omg)
  
# Eulers metode
for i in range (N):
  y[i+1] = y[i] + yder(y[i],x[i])*dx
  x[i+1] = x[i] + dx
  
plot(x,y)
grid()
show()
  
```

```python
from pylab import *

# Fysiske st√∏rrelser
g = 9.81    #tyngeakselerasjonen i m/s^2
m  = 0.5   #Masse til objektet
k = 0.2   #luftmotsatnd til objektet

#Tidsintervaller
N = 10000000    # antall situasjoner
tid = 2   #antall sekunder
dt = tid/(N)   #tidssteg

#Matriser
a = zeros(N)
v = zeros(N)
s = zeros(N)
t = zeros(N)

#initialbetingelser
v[0] = 0
s[0] = 0

# Eulers metoder
for i in range(N-1):
    a[i] = g-(k*v[i]**2)/2  #likning for akselerasjonen
    v[i+1] = v[i]+ a[i]*dt
    s[i+1] = s[i] + v[i]*dt
    t[i+1] = t[i] + dt

#plotting
plot(t,a,"b-")
title("akselereasjon")
xlabel("tid [s]")
ylabel("akselerasjon")

plot(t,v,"r-")
title("Fart")
xlabel("tid [s]")
ylabel("Fart")

plot(t,s,"b-")
title("posisjon")
xlabel("tid [s]")
ylabel("strekning")

legend()
show()
```
