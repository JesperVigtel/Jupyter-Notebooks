### 9.4: 
1) det finnes ikke et nullpunkt til grafen. Den er alltid over x-aksen. 
2) Det er en a-verdi som fører til at den deriverte er lik 0. dette gjør at man deler på 0, noe som "ikke er lov" 
3) man velger 1 som startpunkt, som f(1)= 0. Intervaller som er valgt er derfor ikke mulig, og den tror alltid at den må opp i x-verdi. 
4) Den deriverte av 1 er lik 0, og det deles derfor på 0 i utregningen. 

```python
def f(x): #den har ikke ett nullpunkt
    return x**2+3
def fder(x):
    return 2*x
n = 10000
tol = 1e-10
i = 0
a = 1
x = a
#Newtons metode:
while i <= n and abs(f(x)) >= tol:
    x = a- f(a)/fder(a)
    a = x
    i += 1
    
print("f(",x,") =", f(x))
```
