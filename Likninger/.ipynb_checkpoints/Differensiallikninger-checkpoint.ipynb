# Differensiallikninger


En differenisallikning er en likning som innegolder den deriverte av en funkjson. 
Eksmpler p√• ette er N2, hvor a skrives som v'. $$ \Sigma F = m* v'$$

$$ f'(x) = \frac{(f(x+\Delta x)-f(x))}/{(\Delta x)}$$

```python
from pylab import *

a = 0        # Startverdi
b = 5        # Sluttverdi
dx = 1E-5   # Steglengde
N = int((b-a)/dx) # Antall intervaller
y0 = 1       # Initialbetingelse

# Matriser
x = zeros(N+1)
y = zeros(N+1)
y[0] = y0

# Den deriverte av funksjonen y eller f(x)
def yder(y,x):
  return 1
  
# Eulers metode
for i in range (N):
  y[i+1] = y[i] + yder(y[i],x[i])*dx
  x[i+1] = x[i] + dx
  
plot(x,y)
grid()
  
```

### Newtons avkj√∏lingsmetode


```python
from pylab import *

a = 0             # Startverdi
b = 20            # Sluttverdi
dx = 1E-5         # Steglengde
N = int((b-a)/dx) # Antall intervaller
y0 = 100          # Initialbetingelse
k = -0.5
T_omg = 20

# Matriser
x = zeros(N+1)
y = zeros(N+1)
y[0] = y0


# Den deriverte av funksjonen y eller f(x)
def yder(y,x):
  return k*(y-T_omg)
  
# Eulers metode
for i in range (N):
  y[i+1] = y[i] + yder(y[i],x[i])*dx
  x[i+1] = x[i] + dx
  
plot(x,y)
grid()
show()
  
```
