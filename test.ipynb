```python
def f(x):
    return x**2+4*x-2
xl = []
yl = []

for x in range(10):
    y = f(x)
    xl.append(x)
    yl.append(y)
    
plot(xl,yl)
show()


```

```python
l = linspace(-3,3,14)
f = 4*l-2
g = l**2+5
h = exp(0.05*l)

plot(l,f, color = "red")
plot(l,g,color = "green")
plot(l,h,color = "hotpink")
show()
```
