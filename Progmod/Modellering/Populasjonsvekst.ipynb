#### Harer:
$$H _{t+1} = H_t + aH_t-bH_t*G_t$$
b er møterate / mordrate.
Kan plusse på bæreevne: $$1-\frac{H_t}{B}$$

#### Gauper:
$$G_ {t+1} = G_t + cG_t*H_t-e*G_t$$
