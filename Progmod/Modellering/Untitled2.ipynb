```python
from pylab import * 

#Startbetingelser
N = 157759  #Befolkning
a = 4/N  #Smittetall
b = 3.5  #Dødsrate
tid = 48 #Antall uker

I = 3  #Start syke
S = N  #Start befolkning
R = 0  #Antall resistente

syke = [I]
mulige =[S]
friske = [R]  #Resistente
t = [0]

for i in range(1,tid):
    S = S-a*S*I
    I = I+a*S*I-b*I
    R = R+b*I
    
    syke.append(I)
    mulige.append(S)
    friske.append(R)
    t.append(i)




plot (t,syke,label ="Syke")
#plot(t,mulige,label = "antall mulige")
plot(t,friske,label="friske")
xlabel("Uker")
ylabel("Populasjon")
legend()
show()
```
