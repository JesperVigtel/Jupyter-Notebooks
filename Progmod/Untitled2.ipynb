```python
from pylab import *

N0 = 1000
N = float(N0)
t = 50
k = 0.3
d = 0
B = 100000

tid = []
pop = []

for i in range(t):
    d = 0.25*sin(i)
    k = k-d
    N = N + N*k*(1-(N/B))
    tid.append(i)
    pop.append(N)

plot(tid,pop)
show()
```

```python
from pylab import *

N0 = 1000
N = float(N0)
t = 30
k = 0.4
B = 10000
i = 0

tid = []
pop = []

while N < 9800:
    N = N + N*k*(1-(N/B))
    tid.append(i)
    pop.append(N)
    i += 1

plot(tid,pop)
show()
```
